### Affichage des valeurs aberrantes avec une fonction et le calcul des quantiles

```python
# création d'une fonction pour trouver les valeurs aberrantes
def valeur(v):
    q1=np.quantile(v, 0.25)
    q3=np.quantile(v, 0.75)
    e1=q3-q1
    li=q1-(e1*1.5)
    ls=q3+(e1*1.5)
    i=list(v.index[(v<li)|(v>ls)])
    val=list(v[i])
    return i, val
```

```python
 #affichage des valeurs aberrantes
valeur_aberrante=valeur(wine["price"])
valeur_aberrante[1]
```

```python
# classement des valeurs aberrantes
valeur_aberrante[1].sort()
print(f" Les valeurs aberrantes sont: ")
print(valeur_aberrante[1])
```

### Affichage des valeurs aberrantes en créant une nouvelle colonne et en triant le dataframe

```python
# création de l'ecart type des prix
w1=wine["price"].quantile(0.25)
w3=wine["price"].quantile(0.75)
ecart=w3-w1
    
```

```python
# création des outliers
out= w1-(ecart*1.5)
out1=w3+(ecart*1.5)
```

```python
# création d'une nouvelle colonne
wine.loc[wine["price"]>out1, "statut"]="valeur aberrante"
wine.loc[wine["price"]<=out1, "statut"]="prix normal"
wine
```

```python
# affichage des valeurs aberrantes
outliers=wine.loc[wine["statut"]=="valeur aberrante"]
outliers[["price"]]
```

```python
#affichage des valeurs du plus petit prix au plus grand 
out=outliers[["price"]].sort_values(by="price", ascending=True)
print(f" Les valeurs aberrantes sont :")
out
```

### Affichage des prix aberrants avec la méthode z_score

```python
# création d'une fonction pour la méthode z_score
outliers1=[]
def detect_outliers(price):
    threshold=2
    mean=np.mean(price)
    std=np.std(price)

    for i in price:
        z_score=(i-mean)/std
        if np.abs(z_score)>threshold:
            outliers1.append(i)
    return outliers1
```

```python
# détection des valeurs pour la colonne price
outlier2=detect_outliers(wine["price"])
print(f" nombres de valeurs aberrantes détectées {len(outlier2)}")
outlier2
```

#### Ici on voit que les valeurs aberrantes sont les points colorés en rouge

#### En conclusion la meilleure méthode pour trouver les valeurs aberrantes sera la méthode des quantiles, qui est une méthode simple ettrès efficace
